```python
import yfinance as yf
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickdf = pd.read_csv("data/stocks.csv", error_bad_lines=False, delimiter=";")

tickdf
```

```python
import json

out = []

for idx, row in tickdf.head(10).iterrows():
    tickname = row.Ticker
    
    obj = yf.Ticker(tickname)
    info = obj.info
    
    try:
        if info["quoteType"] == "EQUITY":
            print(tickname)
            out.append(info)
    except:
        continue
        
with open('raw.json', 'w') as f:
    json.dump(out, f)
```

## Load to df

```python
df = pd.read_json('raw.json')    
```
